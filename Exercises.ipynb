```python
#Triangle
sides = [3,3,3]

def isTriangle(sides):
    a = sides[0]
    b = sides[1]
    c = sides[2]
    if a + b > c and a + c > b and b + c > a:
        return True
    return False

def equilateral(sides):
    a, b, c = sides
    if isTriangle(sides):
            if (a == b == c):
                return True
    return False

def isosceles(sides):
    a, b, c = sides
    if isTriangle(sides):
        if (a == b or b == c or c == a):
            return True
    return False

def scalene(sides):
    a, b, c = sides
    if isTriangle(sides):
        if (a != b and b != c and a != c):
            return True
    return False

print(equilateral(sides))
```

```python
#Armstrong Number - PROGRAM
def is_armstrong_number(number):
    #Extrae los dígitos, primero combierte en String el número, luego recorre cada      letra en un ciclo for y la almacena en una lista.
    digits = [int(i) for i in str(number)]
    
    #len(), calcula la cantidad de elementos en la lista
    num_digits = len(digits)
    sum = 0
    
    #print(digits)
    #Se realiza una iteración para calcular la suma de cada numero elevado a la cantidad de dígitos.
    for i in range (0, num_digits):
        sum += (digits[i]**num_digits)
        #print (sum)
    
    #Si suma es igual al número retorna True.
    if sum == number:
        return True
    return False, sum

is_armstrong_number(33)
```

```python
#BlackJack - PROGRAM
letters = ['J','Q','K']

def value_of_card(card):
    """Determine the scoring value of a card.

    :param card: str - given card.
    :return: int - value of a given card.  See below for values.

    1.  'J', 'Q', or 'K' (otherwise known as "face cards") = 10
    2.  'A' (ace card) = 1
    3.  '2' - '10' = numerical value.
    """
    
    card = card.upper()

    if card in letters: card = 10

    if card == 'A': card = 1

    return int(card)


def higher_card(card_one, card_two):
    """Determine which card has a higher value in the hand.

    :param card_one, card_two: str - cards dealt in hand.  See below for values.
    :return: str or tuple - resulting Tuple contains both cards if they are of equal value.

    1.  'J', 'Q', or 'K' (otherwise known as "face cards") = 10
    2.  'A' (ace card) = 1
    3.  '2' - '10' = numerical value.
    """
    card_one_num = value_of_card(card_one)
    card_two_num = value_of_card(card_two)

    if card_one_num > card_two_num: return card_one
    elif card_two_num > card_one_num: return card_two
    
    return card_one, card_two


def value_of_ace(card_one, card_two):
    """Calculate the most advantageous value for the ace card.

    :param card_one, card_two: str - card dealt. See below for values.
    :return: int - either 1 or 11 value of the upcoming ace card.

    1.  'J', 'Q', or 'K' (otherwise known as "face cards") = 10
    2.  'A' (ace card) = 11 (if already in hand)
    3.  '2' - '10' = numerical value.
    """
    card_one_num = value_of_card(card_one)
    card_two_num = value_of_card(card_two)
    card_one = card_one.upper()
    card_two = card_two.upper()
    
    if card_one == 'A' or card_two == 'A':
        return 1
    
    if card_one_num + card_two_num <= 10:
        return 11
    
    if card_one_num + card_two_num > 10:
        return 1


def is_blackjack(card_one, card_two):
    """Determine if the hand is a 'natural' or 'blackjack'.

    :param card_one, card_two: str - card dealt. See below for values.
    :return: bool - is the hand is a blackjack (two cards worth 21).

    1.  'J', 'Q', or 'K' (otherwise known as "face cards") = 10
    2.  'A' (ace card) = 11 (if already in hand)
    3.  '2' - '10' = numerical value.
    """

    card_one_num = value_of_card(card_one)
    card_two_num = value_of_card(card_two)
    card_one = card_one.upper()
    card_two = card_two.upper()
    
    if (card_one == 'A' or card_two == 'A') and (card_one_num == 10 or card_two_num == 10):
        return True
    
    if card_one_num + card_two_num == 21:
        return True

    return False


def can_split_pairs(card_one, card_two):
    """Determine if a player can split their hand into two hands.

    :param card_one, card_two: str - cards dealt.
    :return: bool - can the hand be split into two pairs? (i.e. cards are of the same value).
    """
    
    card_one_num = value_of_card(card_one)
    card_two_num = value_of_card(card_two)
    card_one = card_one.upper()
    card_two = card_two.upper()

    if card_one_num == card_two_num:
        return True
    
    return False


def can_double_down(card_one, card_two):
    """Determine if a blackjack player can place a double down bet.

    :param card_one, card_two: str - first and second cards in hand.
    :return: bool - can the hand can be doubled down? (i.e. totals 9, 10 or 11 points).
    """
    card_one_num = value_of_card(card_one)
    card_two_num = value_of_card(card_two)
    card_one = card_one.upper()
    card_two = card_two.upper()
    sum = card_one_num + card_two_num
    
    if (9 <= sum <= 11):
        return True
    
    return False
 
```

```python
#Perfect Numbers
'''
Determine if a number is perfect, abundant, or deficient based on Nicomachus' (60 - 120 CE) classification scheme for positive integers.

The Greek mathematician Nicomachus devised a classification scheme for positive integers, identifying each as belonging uniquely to the categories of perfect, abundant, or deficient based on their aliquot sum. The aliquot sum is defined as the sum of the factors of a number not including the number itself. For example, the aliquot sum of 15 is 1 + 3 + 5 = 9.

Perfect
A number is perfect when it equals its aliquot sum. For example:

6 is a perfect number because 1 + 2 + 3 = 6
28 is a perfect number because 1 + 2 + 4 + 7 + 14 = 28

Abundant
A number is abundant when it is less than its aliquot sum. For example:

12 is an abundant number because 1 + 2 + 3 + 4 + 6 = 16
24 is an abundant number because 1 + 2 + 3 + 4 + 6 + 8 + 12 = 36

Deficient
A number is deficient when it is greater than its aliquot sum. For example:

8 is a deficient number because 1 + 2 + 4 = 7
Prime numbers are deficient

Task
Implement a way to determine whether a given number is perfect. Depending on your language track, you may also need to implement a way to determine whether a given number is abundant or deficient.


'''

def classify(number):
    """ A perfect number equals the sum of its positive divisors.

    :param number: int a positive integer
    :return: str the classification of the input integer
    """
    #1. Tomar el parametro number y recorrer del 1 al number para determinar que numeros son múltiplos.
    #2. Los numeros que sea múltiplos almacenarlos en una lista.
    #3. Sumar cada numero de la lista.
    #4. Si el resultado de la suma es igual a number, entonces es perfecto.
    
    factors_nums = []
    
    if number <= 0:
        raise ValueError("Classification is only possible for positive integers.")
    
    for num in range(1, number):
        if number % num == 0:
            factors_nums.append(num)
    
    
    aliquot_sum = sum(factors_nums)
    
    if aliquot_sum == number:
        return "Perfect"
    elif aliquot_sum > number:
        return "Abundant"
    else: return "Deficient"
        

classify(6) 
```

```python
#Sublists
'''
Given any two lists A and B, determine if:

List A is equal to list B; or
List A contains list B (A is a superlist of B); or
List A is contained by list B (A is a sublist of B); or
None of the above is true, thus lists A and B are unequal

Specifically, list A is equal to list B if both lists have the same values in the same order. List A is a superlist of B if A contains a contiguous sub-sequence of values equal to B. List A is a sublist of B if B contains a contiguous sub-sequence of values equal to A.

Examples:

If A = [] and B = [] (both lists are empty), then A and B are equal
If A = [1, 2, 3] and B = [], then A is a superlist of B
If A = [] and B = [1, 2, 3], then A is a sublist of B
If A = [1, 2, 3] and B = [1, 2, 3, 4, 5], then A is a sublist of B
If A = [3, 4, 5] and B = [1, 2, 3, 4, 5], then A is a sublist of B
If A = [3, 4] and B = [1, 2, 3, 4, 5], then A is a sublist of B
If A = [1, 2, 3] and B = [1, 2, 3], then A and B are equal
If A = [1, 2, 3, 4, 5] and B = [2, 3, 4], then A is a superlist of B
If A = [1, 2, 4] and B = [1, 2, 3, 4, 5], then A and B are unequal
If A = [1, 2, 3] and B = [1, 3, 2], then A and B are unequal
'''

# Possible sublist categories.
# Change the values as you see fit.
SUBLIST = "SUBLIST"
SUPERLIST = "SUPERLIST"
EQUAL = "EQUAL"
UNEQUAL = "UNEQUAL"

def sublist(list_one, list_two):
    """
    Determina la relación entre dos listas.
    
    Retorna:
    - EQUAL: si las listas son idénticas
    - SUBLIST: si list_one está contenida en list_two
    - SUPERLIST: si list_two está contenida en list_one
    - UNEQUAL: si no tienen relación
    """
    
    # Caso 1: ¿Son idénticas las listas?
    if list_one == list_two:
        return EQUAL
    
    # Caso 2: ¿list_one está vacía?
    # Una lista vacía siempre es sublist de cualquier otra
    if len(list_one) == 0:
        return SUBLIST
    
    # Caso 3: ¿list_two está vacía?
    # Si list_two está vacía y list_one no, entonces list_one es superlist
    if len(list_two) == 0:
        return SUPERLIST
    
    # Caso 4: Buscar si list_one está dentro de list_two
    # Solo tiene sentido si list_one es más pequeña o igual
    if len(list_one) <= len(list_two):
        # Recorre cada posición posible en list_two
        for i in range(len(list_two) - len(list_one) + 1):
            # Extrae un "pedazo" de list_two del mismo tamaño que list_one
            # Ejemplo: list_two[0:3], list_two[1:4], list_two[2:5], etc.
            if list_two[i:i+len(list_one)] == list_one:
                return SUBLIST
    
    # Caso 5: Buscar si list_two está dentro de list_one
    # Solo tiene sentido si list_two es más pequeña
    if len(list_two) < len(list_one):
        # Recorre cada posición posible en list_one
        for i in range(len(list_one) - len(list_two) + 1):
            # Extrae un "pedazo" de list_one del mismo tamaño que list_two
            if list_one[i:i+len(list_two)] == list_two:
                return SUPERLIST
    
    # Si ningún caso anterior se cumple, las listas no tienen relación
    return UNEQUAL
    
```

```python
#Little Sister's Vocabulary

def add_prefix_un(word):
    """Take the given word and add the 'un' prefix.

    :param word: str - containing the root word.
    :return: str - of root word prepended with 'un'.
    """
    return "un" + word

def make_word_groups(vocab_words):
    """Transform a list containing a prefix and words into a string with the prefix followed by the words with prefix prepended.

    :param vocab_words: list - of vocabulary words with prefix in first index.
    :return: str - of prefix followed by vocabulary words with
            prefix applied.

    This function takes a `vocab_words` list and returns a string
    with the prefix and the words with prefix applied, separated
     by ' :: '.

    For example: list('en', 'close', 'joy', 'lighten'),
    produces the following string: 'en :: enclose :: enjoy :: enlighten'.
    """
    vocab_words =[vocab_words[0]] + [vocab_words[0] + word for word in vocab_words[1:]]
    return " :: ".join(vocab_words)

def remove_suffix_ness(word):
    """Remove the suffix from the word while keeping spelling in mind.

    :param word: str - of word to remove suffix from.
    :return: str - of word with suffix removed & spelling adjusted.

    For example: "heaviness" becomes "heavy", but "sadness" becomes "sad".
    """
    word = word[:-4]
    
    if word[len(word)-1] == "i":
        return word.replace("i", "y")
    
    return word



def adjective_to_verb(sentence, index):
    """Change the adjective within the sentence to a verb.

    :param sentence: str - that uses the word in sentence.
    :param index: int - index of the word to remove and transform.
    :return: str - word that changes the extracted adjective to a verb.

    For example, ("It got dark as the sun set.", 2) becomes "darken".
    """
    return sentence.split(".")[0].split()[index] + "en"

sentence ="Look at the bright sky."
index = -2
adjective_to_verb(sentence,index)

    
```

```python
#Pangram
def is_pangram(sentence):
    ALPHABET = ("a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z")
    sentence = sentence.strip().lower().replace(" ","")
    for index in ALPHABET:
        if index not in sentence: return False
    return True
    
is_pangram("quick brown fox jumps over the lazy dog")
```

```python
#Isograms
'''Determine if a word or phrase is an isogram.

An isogram (also known as a "non-pattern word") is a word or phrase without a repeating letter, however spaces and hyphens are allowed to appear multiple times.

Examples of isograms:

lumberjacks
background
downstream
six-year-old'''

def is_isogram(string):
    string = string.lower().strip().replace("-","").replace(" ","")
    for index in string:
        count = string.count(index)
        if count > 1: return False
    return True
        
        
is_isogram("A2lFons--")
```

```python
#Start ISBN Verifier
'''The ISBN-10 verification process is used to validate book identification numbers. These normally contain dashes and look like: 3-598-21508-8

ISBN
The ISBN-10 format is 9 digits (0 to 9) plus one check character (either a digit or an X only). In the case the check character is an X, this represents the value '10'. These may be communicated with or without hyphens, and can be checked for their validity by the following formula:

(d₁ * 10 + d₂ * 9 + d₃ * 8 + d₄ * 7 + d₅ * 6 + d₆ * 5 + d₇ * 4 + d₈ * 3 + d₉ * 2 + d₁₀ * 1) mod 11 == 0

If the result is 0, then it is a valid ISBN-10, otherwise it is invalid.
'''

def is_valid(isbn):
    isbn = isbn.replace("-","")
    if len(isbn) != 10: return False
    sum = 0
    count = 10
    if int(isbn.isnumeric()) or isbn.endswith("X"):
        while count > 0:
            for index in isbn:
                #print(f"index: {index}, count:{count}, ope: {int(index)*count}, sum:{sum} ")
                if index == "X": sum = sum + (10 * 1)
                elif index.isalpha():
                    return False
                else: 
                    sum = sum + (int(index) * count)
                count = count - 1
        return True if sum % 11 == 0 else False
    else:
        return False    
    
    
is_valid("3-598-P1581-X")
```

```python
#Rotational Cipher
''' Create an implementation of the rotational cipher, also sometimes called the Caesar cipher.

The Caesar cipher is a simple shift cipher that relies on transposing all the letters in the alphabet using an integer key between 0 and 26. Using a key of 0 or 26 will always yield the same output due to modular arithmetic. The letter is shifted for as many values as the value of the key.

The general notation for rotational ciphers is ROT + <key>. The most commonly used rotational cipher is ROT13.

A ROT13 on the Latin alphabet would be as follows:

Plain:  abcdefghijklmnopqrstuvwxyz
Cipher: nopqrstuvwxyzabcdefghijklm

Examples
ROT5 omg gives trl
ROT0 c gives c
ROT26 Cool gives Cool
ROT13 The quick brown fox jumps over the lazy dog. gives Gur dhvpx oebja sbk whzcf bire gur ynml qbt.
ROT13 Gur dhvpx oebja sbk whzcf bire gur ynml qbt. gives The quick brown fox jumps over the lazy dog.

'''

def rotate(text, key):
    ALPHABET = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    
    #Se Valida si la llave es diferente a cero y crea una nueva lista con el alfabeto rotado
    if key != 0:    
        rotated_alphabet = ALPHABET[key:]
        rotated_alphabet.extend(ALPHABET[:key])
    else: rotated_alphabet = ALPHABET
    
    #Se Crea una variable tipo string vacía donde se concatenarán los caracteres
    rotated_text = ""
    
    #Se recorre el texto caracter por caracter y se valida si se encuentra dentro del alfabeto o no.
    for character in text:    
        if character.lower() in ALPHABET:
            if character.isupper():
                rotated_text = rotated_text + rotated_alphabet[ALPHABET.index(character.lower())].upper()
            else:
                rotated_text = rotated_text + rotated_alphabet[ALPHABET.index(character.lower())]
        else: rotated_text = rotated_text + character
    
    return rotated_text
    
        
    
rotate("Luis Porto 33, Jorge Arroyo 33, Alfonso 32", 5)
```

```python
#Card Games - PROGRAM

"""Functions for tracking poker hands and assorted card tasks."""


def get_rounds(number):
    """Create a list containing the current and next two round numbers.

    :param number: int - current round number.
    :return: list - current round and the two that follow.
    """    
    list = [number, number+1, number+2]
    
    return list


def concatenate_rounds(rounds_1, rounds_2):
    """Concatenate two lists of round numbers.

    :param rounds_1: list - first rounds played.
    :param rounds_2: list - second set of rounds played.
    :return: list - all rounds played.
    """
    
    return rounds_1 + rounds_2


def list_contains_round(rounds, number):
    """Check if the list of rounds contains the specified number.

    :param rounds: list - rounds played.
    :param number: int - round number.
    :return: bool - was the round played?
    """

    return True if number in rounds else False


def card_average(hand):
    """Calculate and returns the average card value from the list.

    :param hand: list - cards in hand.
    :return: float - average value of the cards in the hand.
    """

    return sum(hand)/len(hand)


def approx_average_is_average(hand):
    """Return if the (average of first and last card values) OR ('middle' card) == calculated average.

    :param hand: list - cards in hand.
    :return: bool - does one of the approximate averages equal the `true average`?
    """
    hand = sorted(hand)
    #print(hand)
    #print(f"hand[0]: {hand[0]}")
    #print(f"hand[len(hand)-1]: {hand[len(hand)-1]}")
    avg_first_last = (hand[0] + hand[len(hand)-1])/2
    #print (f"avg_first_last: {avg_first_last}")
    avg_middle = hand[int((len(hand)-1)/2)]
    #print(f"avg_middle: {avg_middle}")
    avg_calc = card_average(hand)
    #print(f"avg_calc: {avg_calc}")
    return True if (avg_first_last == avg_calc or avg_middle == avg_calc) else False


def average_even_is_average_odd(hand):
    """Return if the (average of even indexed card values) == (average of odd indexed card values).

    :param hand: list - cards in hand.
    :return: bool - are even and odd averages equal?
    """
    hand = sorted(hand)
    odd_list = []
    even_list = []
    for index in range (len(hand)):
        if index % 2 == 0:
            odd_list.append(hand[index])
        else:
            even_list.append(hand[index])
    return True if card_average(odd_list) == card_average(even_list) else False


def maybe_double_last(hand):
    """Multiply a Jack card value in the last index position by 2.

    :param hand: list - cards in hand.
    :return: list - hand with Jacks (if present) value doubled.
    """
    last_position = len(hand)-1
    
    if hand[last_position] == 11:
        hand[last_position] = hand[last_position]*2
    
    return hand


get_rounds(2)
concatenate_rounds(get_rounds(2), get_rounds(10))
list_contains_round(get_rounds(5), 7)
card_average([3,8,12])
approx_average_is_average([1,2,3,4,2])
average_even_is_average_odd([2,2,2,2])
maybe_double_last([1,2,11])
```
